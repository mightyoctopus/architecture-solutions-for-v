## Prototype Code 

Note: This demo code was designed for prototyping purposes. Specific implementations and bolierplate code are skipped (e.g. LLM API calls, prompt formatting, retry logic etc)

### Mockup snapshot context

```python
### Mockup data ###
kernel_snapshot = {
    "active_task": "Continuous Discovery",
    "current_step": 1,
    "sprint_priority": "high",
    "behavior_pattern": "slow_execution",
}
```

### Mockup ObserverContext Layer

```python
class ObserverContext:
    def __init__(self, kernel_snapshot: dict):
        self.kernel_snapshot = kernel_snapshot

    def get_context(self):
        return self.kernel_snapshot
```

### Mockup ObserverReasoning Layer

```python
class ObserverReasoning:
    def __init__(self, context: ObserverContext):
        self.context = context.get_context()

    def generate_reasoning(self):
        """
        Shared reasoning generated ONCE centrally
        """

        return {
            "guidance_mode": {
                "signal": "task_stalling",
                "priority": "medium",
                "time_elapsed_minutes": 20,
                "expected_duration_minutes": 8,
                "tone": "supportive",
                "recommended_action": "encourage_reflection"
            },

            "notification_bubble": {
                "signal": "goal_inactive",
                "priority": "urgent",
                "inactive_days": 10,
                "tone": "emotional_urgency",
                "recommended_action": "check_goal_status"
            },

            "daily_brief": {
                "signal": "momentum_drop",
                "priority": "low",
                "tone": "reflective_honest",
                "recommended_action": "summarize_execution_pattern"
            }
        }
```

### Mockup Renderer Layers

```python
class GuidanceModeRenderer: 
    def render(self, reasoning_data: dict):
        return (
            f"Twenty minutes on this already. "
            f"Estimate was {reasoning_data['expected_duration_minutes']}. "
            f"Stuck or just thorough?"
        )


class NotificationBubbleRenderer:
    def render(self, reasoning_data: dict):
        return (
            f"Goal has been inactive for "
            f"{reasoning_data["inactive_days"]} days"
        )
        

class DailyBriefRenderer:
    def render(self, reasoning_data: dict):
        return (
            "Execution pacing slowed down today. "
            "Might be worth reviewing task focus."
        )
```

### Mockup Controler Layer

```python
# Shared Observer Context 
observer_context = ObserverContext(kernel_snapshot)


# Centralized Reasoning
observer_reasoning = ObserverReasoning(context=observer_context)
reasoning_output = observer_reasoning.generate_reasoning()


# Surface renderers
guidance_renderer = GuidanceModeRenderer()
notification_renderer = NotificationBubbleRenderer()
brief_renderer = DailyBriefRenderer()


# Render final outputs
guidance_output = guidance_renderer.render(
    reasoning_output["guidance_mode"]
)

notification_output = notification_renderer.render(
    reasoning_output["notification_bubble"]
)

brief_output = brief_renderer.render(
    reasoning_output["daily_brief"]
)
```

### Mockup Output Examples

```python
print("\n############## Guidance Mode ##############")
print(guidance_output)

print("\n\n############## Notification Bubble ##############")
print(notification_output)

print("\n\n############## Daily Brief ##############")
print(brief_output)
```
